## (2 points) 1B: Bayes' Rule (discrete random variables)
Let $X\in\{T,F\}$ be a binary random variable indicating whether a patient has a disease ($T$) or not ($F$). The (marginal or unconditional) probability of a patient having the disease is 0.01. Sick patients test positive for the disease 90% of the time. However, healthy patients test positive 5% of the time. 

1. What is the probability of having the disease if you test positive?

### <span style="color:blue">Answer</span>
$p(X = T \, \mid \, tested \, positive)$

$= \dfrac{p(X = T, tested \, positive)}{p(tested \, positive)}$ # the product rule

$= \dfrac{p(X = T, tested \, positive)}{p(tested \, positive, X = F) + p(tested \, positive, X = T)}$ # the sum rule

$= \dfrac{p(tested \, positive \,\mid\, X = T)p(X=T)}{p(tested \, positive \, \mid \, X = F)p(X = F) + p(tested \, positive \, \mid \, X = T)p(X=T)}$ # the product rule

$= \dfrac{0.01 \times 0.9}{0.05 \times 0.99 + 0.9 \times 0.01}$

$= 0.154$

## (2 points) 1C: Bayes' Rule (continuous random variables)
A croupier samples a standard normal $X\sim \mathcal{N}(\mu=0,\sigma^2=1)$, and tosses a fair coin. If the coin lands on heads, she reveals to you the number $Z=X+1$. If it comes up tails she reveals $Z=X-1$. 

1. What is the probability that the number she reveals, $Z$, is greater than $a$, given that she tossed heads? 
- What is the probability she tossed heads, if the number she reveals to you is greater than $a$? 

Express the results in terms of the cumulative distribution function of the standard normal, that is $\Phi(a):=P(X<a)$.

### <span style="color:blue">Answer</span>
<i>1. $p(Z > a \mid head) $</i>

<i>$ = \dfrac{p(Z > a, head)}{p(head)}$</i> # the product rule

<i>$ = \dfrac{p(X > a-1)p(head)}{p(head)}$</i> # from known conditions

<i>$ = 1-P(X<a-1)$</i>

<i>$ = 1-\Phi(a-1)$</i> # from known conditions

<i>2. $p(head \mid Z > a) $</i>

<i>$ = \dfrac{p(Z > a, head)}{p(Z > a)}$</i> # the product rule

<i>$ = \dfrac{p(Z > a, head)}{p(Z > a, head) + p(Z > a, tail)}$</i>　＃the sum rule

<i>$ = \dfrac{p(X > a-1)p(head)}{p(X > a -1)p(head) + p(X > a + 1)p(tail)}$</i> # from known conditions

<i>$ = \dfrac{1-\Phi(a-1)}{1-\Phi(a-1) + 1-\Phi(a+1)}$</i> # from known conditions

<i>$ =  \dfrac{1-\Phi(a-1)}{２-\Phi(a-1)-\Phi(a+1)}　$</i> # simplify

## (5 points) 1D: Bayes Rule Simulation and Plot
1. Write a function ```theoretical_p``` which takes argument $a$ and returns the theoretical conditional probability derived previously, using ```scipy.stats.norm.cdf``` to compute $\Phi$
- Write a function ```monte_carlo_p``` which takes arguments $a$ and $n$, and which performs the croupier's sampling procedure $n$ times and returns an empirical estimate of the probability we computed above. Use the functions ```np.random.randn```, ```np.random.binomial``` and ```np.mean```. Avoid explicit loops, instead rely on the boolean index array feature of numpy.
- Plot the theoretical solution as a function of $a\in [-5,5]$ using the functions ```np.linspace``` to make a grid $1024$ of values for $a$.
- Overlay a plot of the empirical estimate with $n=500$. Label the two plots using the ```label``` argument to ```plt.plot``` along with ```plt.legend(loc='best')``` after plotting. Always label the axes in your plots. Add grid lines with ```plt.grid()```. Matplotlib accepts latex strings such as ```xlabel=r'$a$'```.

Think about the intuition behind the limiting values of the plot as $a\rightarrow-\infty$ and $a\rightarrow+\infty$.

````python
## Note: import packages before running this cell

## Question 1
def theoretical_p(a):
    
    # formula derived in the subquestion 1 of the question 1C
    p_1c_1 = 1-norm.cdf(a-1)  
    
    # formula derived in the subquestion 2 of the question 1C
    p_1c_2 = (1-norm.cdf(a-1))/(2 - norm.cdf(a-1) - norm.cdf(a+1))
    
    return(p_1c_1, p_1c_2)
    

## Question 2
def monte_carlo_p(a,n):
    
    # n trails
    trails = np.random.binomial(1,0.5,size = n)

    # n samples
    samples_from_normal_distribuion = np.random.randn(n)
    
    # n revealed numbers
    revealed_numbers = samples_from_normal_distribuion - 1 + 2*trails
    
    # formula derived in the subquestion 1 of the question 1C
    revealed_number_given_heads = revealed_numbers[np.where(trails > 0)]
    p_1c_1 = np.mean(revealed_number_given_heads>a)
    
    # formula derived in the subquestion 2 of the question 1C
    trails_given_revealed_numbers_greater_than_a = trails[np.where(revealed_numbers>a)]
    p_1c_2 = np.mean(trails_given_revealed_numbers_greater_than_a)
    
    return(p_1c_1, p_1c_2)


## Question 3 & 4
n= 500
a_array = np.linspace(-5,5,1024)
theoretical_ps = [theoretical_p(a) for a in a_array]
empirical_ps = [monte_carlo_p(a,n) for a in a_array]

fig = plt.figure(figsize=(11,8))

# Draw figure1
ax = fig.add_subplot(211)
ax.plot(a_array, [p[0] for p in theoretical_ps], label = "theoretical")
ax.plot(a_array, [p[0] for p in empirical_ps], 'r.',label = "empirical")
plt.legend(loc ='best')
ax.set_ylabel('probability')
ax.set_title('Theoretical and empirical probabilities of subquestion 1 on [-5, 5]')
ax.set_xlabel('$a$')
plt.grid()

# Draw figure2
ax = fig.add_subplot(212)
ax.plot(a_array, [p[1] for p in theoretical_ps], label = "theoretical")
ax.plot(a_array, [p[1] for p in empirical_ps],'r.' ,label = "empirical")
plt.legend(loc ='best')
ax.set_ylabel('probability')
ax.set_xlabel('$a$')
ax.set_title('Theoretical and empirical probabilities of subquestion 2 on [-5, 5]')
plt.grid()
````

## (2 Points) 2A: Linear Regression (theory)
Assume the deterministic model $y_i=\sum_{j=1}^d w_j x_{i,j}, i=1, 2, \ldots n$.
1. Write the model in matrix form using capital letters for matrices and bold lowercase for vectors, stating the dimensionality of the variables.
- Assume we observe corrupted targets $\hat{y}_i=y_i+\epsilon_i$, where $\epsilon_i$ represents unpredictable noise. Write the sum of squared errors $\sum_{i=1}^n (\sum_{j=1}^d w_j x_{i,j}-\hat{y}_i)^2$ in matrix form, and derive the vector $\mathbf{w}^\star$ which minimises it. You may wish to google <i>e.g.</i> "Sam Roweis matrix identities" and look at the section on "derivatives of scalar forms".

### <span style="color:blue">Answer</span>
1. \begin{equation}\mathbf{y} = X \mathbf{w},\end{equation}
where $\mathbf{y}$ is a n-dimentional column vector, $\mathbf{w}$ is a d-dimensional column vector and $X$ is a $n\times d$ matrix.
- (1) 
\begin{equation}
\sum_{i=1}^n (\sum_{j=1}^d w_j x_{i,j}-\hat{y}_i)^2 = (X\mathbf{w} - \hat{\mathbf{y}})^{T}(X\mathbf{w} - \hat{\mathbf{y}}). 
\end{equation}

  (2) In order to get the vector $\mathbf{w}^\star$ which minimises the error, first we compute the partial derivatives of errors with respect to $\mathbf{w}$ in the direction $\mathbf{\xi}$.
   \begin{equation}
    D[(X\mathbf{w} - \hat{\mathbf{y}})^{T}(X\mathbf{w} - \hat{\mathbf{y}})](\mathbf{\xi})=2\mathbf{\xi}^T (X^{T}X\mathbf{w}-X^{T}\hat{\mathbf{y}})
\end{equation}
  Make the partial derivatives zero in all directions and we get $\mathbf{w}^\star$:
  $$\mathbf{w}^\star = (X^{T}X)^{-1}X^{T} \hat{\mathbf{y}}$$
